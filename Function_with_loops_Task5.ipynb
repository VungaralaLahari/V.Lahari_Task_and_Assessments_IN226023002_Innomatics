# Problem 1: Social Media – Post Engagement Analyzer
## Problem Statement
Create a function to analyze engagement on a social media post.
## Rules
- Calculate total likes
- If Total Likes ≥ 1000 → Viral Post
- Otherwise → Normal Engagement


```python
def post_engagement_analyser(likes):
    total_likes=0
    for like in likes:
        total_likes+=like
    print("Total Likes:",total_likes)
    if total_likes>=1000:
        return "Post Status: Viral Post"
    else:
        return "Post Status: Normal Engagement"
list_likes=[1022,234,567,1000,2000]
print(post_engagement_analyser(list_likes))
```

# Problem 3: Agriculture – Rainfall Adequacy Checker
## Problem Statement
Create a function that analyzes seasonal rainfall data.
## Rules
- Calculate average rainfall
- Average ≥ required level → Adequate Rainfall
- Otherwise → Inadequate Rainfall


```python
def rainfall_checker(rainfall):
    total=0
    required_level=70
    for rain in rainfall:
        total+=rain
    average_rainfall=total/len(rainfall)
    print("Average Rainfall:",average_rainfall)
    if average_rainfall>=required_level:
        print("Rainfall Status: Adequate Rainfall")
    else:
        print("Rainfall Status: Inadequate Rainfall")
rainfall=[60,75,85,90,91]
rainfall_checker(rainfall)
        
```

# Problem 4: Social Media – Duplicate Account Detection
## Problem Statement
Create a function to detect duplicate usernames in a system.
## Rules
- If duplicates exist → Flag accounts
- Else → Accounts are unique


```python
existing_users=["rahul","karan","arjun","sunitha"]
def duplicate_detection(user_list):
    
    if len(user_list) !=len(set(user_list)):
        print("Duplicate Accounts Found: Yes")
    else:
        print("Duplicate Accounts Found: No")
users_list=input("Enter user name :").split()
all_users=users_list+existing_users
duplicate_detection(all_users)

```

# Problem 6: Agriculture – Premium Crop Price Filter
## Problem Statement
Create a function that filters crop prices above a premium threshold.
## Rules
- Price > ₹2000 → Premium Crop
- Else → Regular Crop


```python
premium_crop=[]
regular_crop=[]
def crop_filter(crops):
    for crop in crops:
        if crop>2000:
            premium_crop.append(crop)
            
        else:
            regular_crop.append(crop)
            

crop=[2000,1500,900,3000,1200]
crop_filter(crop)
print("Premium Crop:",premium_crop)
print("Regular Crop:",regular_crop)
```

# Problem 7: System Monitoring – Application Health Checker
## Problem Statement
Create a function to evaluate application health based on error count.
## Rules
- Errors = 0 → Healthy
- Errors ≤ 5 → Minor Issues
- Errors > 5 → Critical Issues


```python
def health_checker(error):
    if error<0:
        print("Invalid")
    elif error ==0:
        print("System Status: Healthy")
    elif error>0 and error<=5:
        print("System Status: Minor Issues")
    else:
        print("System Status: Critical Issues ")

error=int(input("Enter Error Count:"))
print("Error Count:",error)
health_checker(error)
        
```

# Problem 8: Banking – Daily Transaction Limit Checker
## Problem Statement
Create a function to validate daily transaction amount.
## Rules
- Daily limit = ₹50,000
- Amount within limit → Approved
- Amount exceeds limit → Rejected


```python
def transaction_checker(daily_transaction):
    daily_limit=50000
    total=0
    for transaction in daily_transaction:
        total+=transaction
    print("Transaction Amount:",total)
    if total<=0:
        print("Transaction Status: NULL")
    elif total <= 50000:
        print("Transaction Status: Approved")
    else:
        print("Transaction Status: Rejected")

todays_transactions=list(map(int,input("Enter Today's Transaction:").split()))
transaction_checker(todays_transactions)
```

# Problem 9: E-Learning – Student Attendance Eligibility System
## Problem Statement
Create a function to calculate attendance percentage and check exam eligibility.
## Rules
- Attendance ≥ 75% → Eligible
- Attendance < 75% → Not Eligible


```python
def attendance_eligibility(attendance):
    present_count=0
    for i in attendance :
        if i == "P":
            present_count+=1
    attendance_percentage=(present_count/len(attendance))*100
    print(f"Attendance Percentage:{attendance_percentage:.2f}")
    if  attendance_percentage>=75.00:
        print("Exam Eligibility: Eligible")
    else:
        print("Exam Eligibility: Not Eligible")

attendance=list(input("Enter Attendance:").split())
attendance_eligibility(attendance)
```

# Problem 10: Smart Electricity Bill Analyzer
## Problem Statement
You are building a smart electricity billing system for residential users.
Create a function that calculates the electricity bill and classifies usage level based on units consumed.
## Billing Rules
- First 100 units → ₹3 per unit
- Next 100 units (101–200) → ₹5 per unit
- Above 200 units → ₹7 per unit
## Usage Classification
- Total bill < ₹500 → Low Usage
- ₹500 – ₹1500 → Moderate Usage
- ₹1500 → High Usage


```python


def bill_analyzer(units):
    total_bills=units
    if units <=100:
        total_bills*=3
    elif units > 100 and units <=200:
        total_bills*=5
    elif units >200:
        total_bills*=7
    else:
        print("Invalid Units")

    print("Total Bill:",total_bills)

    if total_bills <=500:
        print("Low Usage")
    elif total_bills>500 and total_bills<=1500:
        print("Moderate Usage")
    else:
        print("High Usage")
        
units=int(input("Enter Units:"))
bill_analyzer(units)
```
