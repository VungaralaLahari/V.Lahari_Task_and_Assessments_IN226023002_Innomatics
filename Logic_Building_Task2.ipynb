## Problem 1: Unique Words in a Sentence


```python
# writing the input data
sentence='python is easy and python is powerful'
# set does not  allow duplicate
unique_words=set(sentence.split())
# output displaying
# length of the set is being printed 
print("Unique words count:",len(unique_words))
print("Unique words:",unique_words)
        
```

## Problem 2: Highest Salary from Employee Data


```python
# Creating a dictionary where key -> employee name and value -> employee salary
employees = {
    "Ravi": 75000,
    "Anita": 68000,
    "Kiran": 72000,
    "Lahari": 90000,
    "Deepak": 45000
}

maximum_employee = max(employees, key=employees.get)

print(f"Highest salary: {maximum_employee} - {employees[maximum_employee]}")
```

## Problem 3: Find Maximum and Minimum Values


```python
# input 
numbers = [45, 22, 89, 10, 66]
# firstly initializing the max and min values to be the first element
maximum=numbers[0]
minimum=numbers[0]
for n in numbers:
    # comparing the values and reassigning
    if n>maximum:
        maximum=n
    if n < minimum:
        minimum=n
# displaying the output 
print("List:",numbers)
print("Max:",maximum)
print("Min:",minimum)
```

## Problem 4: Count Products Above a Price Threshold

```python
# input
prices = [450, 1200, 899, 1500, 300]
# the count of price above 1000 is taken as 0 at start 
count=0
# iterating over the list
for price in prices:
    # checking condition
    if price>1000:
        count+=1
        # increasing the count
# displaying output
print("Products above 1000:",count)
```

## Problem 5: Calculate Attendance Percentage


```python
# attendance input 
attendance = ["P", "P", "A", "P", "P"]
# intial present count is assigned as 0
present_count=0
for i in attendance:
    if i =='P':
        present_count+=1
# calculating attendance
attendance_percent=(present_count/len(attendance))*100
print("Attendance Percentage:",attendance_percent)
```

## Problem 7: Count Character Frequency


```python
# input string 
s="programming"
# empty dictionary to store character counts
freq={}
# iterate through each character in string
for ch in s:
    # if character not in freq{} we add it to the dictionary and give count 1
    if ch not in freq:
        freq[ch]=1
    else:
        # if character already in freq{} we increase the frequency count 
        freq[ch]+=1
print(freq)
```

## Problem 9: Check if a Key Exists in a Dictionary

## Problem 10: Calculate Average Marks

```python
# sample input marks
marks = [65, 70, 80, 75]
# adding all marks
total = sum(marks)
# finding the average of marks
average = total / len(marks)
# displaying the output 
print("Average Marks:", average)
```
