# Problem 1: Smart Light Controller
## Problem Statement
Create a class to control a smart
light that can be turned ON or OFF and display its status.


```python
# class for Smart Light
class SmartLight:
    
    # To set light name and default status
    def __init__(self, name):
        self.name = name
        self.status = "OFF"   # light is OFF by default

    #  to turn light ON
    def turn_on(self):
        self.status = "ON"

    #  to turn light OFF
    def turn_off(self):
        self.status = "OFF"

    # to display current status
    def show_status(self):
        print(self.name + " is " + self.status)


# Taking input from user
light_name = input("Light Name: ")
action = input("Action (ON/OFF): ")

# Create object of SmartLight
light = SmartLight(light_name)

# Perform action based on input
if action == "ON":
    light.turn_on()
elif action == "OFF":
    light.turn_off()

# Display 
light.show_status()
```

# Problem 2: Employee ID Card System
## Problem Statement
Create a class that stores employee name, ID, and department and displays ID card details.



```python
# Create a class for Employee ID Card
class Employee:
    # this will store employee details
    def __init__(self, name, emp_id, department):
        self.name = name          
        self.emp_id = emp_id     
        self.department = department   

    # method to display ID card details
    def display_id_card(self):
        print("Employee ID Card")
        print("Name:", self.name)
        print("ID:", self.emp_id)
        print("Department:", self.department)


# taking input from user
name = input("Employee Name: ")          
emp_id = input("Employee ID: ")          
department = input("Department: ")       


# creating object of Employee class
emp = Employee(name, emp_id, department)

# calling function to display ID card
emp.display_id_card()
```

# Problem 3: Mobile Contact Record
## Problem Statement
Create a class to store contact name and phone number and display contact information.


```python
# Create a class for Mobile Contact
class Contact:
    # this will store contact details
    def __init__(self, name, phone):
        self.name = name      
        self.phone = phone    

    # method to display contact details
    def display_contact(self):
        print("Contact Saved")
        print("Name:", self.name)
        print("Phone:", self.phone)


# taking input from user
name = input("Contact Name: ")        
phone = input("Phone Number: ")       
# creating object 
c1 = Contact(name, phone)
# calling method to display contact information
c1.display_contact()
```

# Problem 4: Product Price Tag Generator
## Problem Statement
Create a class that stores product name and price and prints a formatted price tag.


```python

class Product:
    # this will store product details
    def __init__(self, name, price):
        self.name = name      
        self.price = price   

    # method to print price tag
    def print_price_tag(self):
        print("Product:", self.name)
        print("Price: ₹" + str(self.price))  

# taking input from user
name = input("Product Name: ")      
price = input("Price: ")            
# creating object 
p1 = Product(name, price)
# calling method 
p1.print_price_tag()
```

## Problem 5: Movie Rating Display System
# Problem Statement
Create a class that stores movie name and rating and displays movie details.


```python
class Movie:

    # this will store movie details
    def __init__(self, name, rating):
        self.name = name        
        self.rating = rating    

    # method to display movie details
    def display_movie(self):
        print("Movie:", self.name)
        print("Rating:", self.rating, "/ 5") 


# taking input from user
name = input("Movie Name: ")      
rating = input("Rating: ")      
# creating object 
m1 = Movie(name, rating)
# calling method 
m1.display_movie()
```

# Problem 6: Delivery Address Manager
## Problem Statement
Create a class that stores customer name and delivery address and prints delivery details.


```python
class Delivery:
    # this will store customer name and address
    def __init__(self, customer_name, address):
        self.customer_name = customer_name   
        self.address = address               

    # method to print delivery details
    def print_details(self):
        print("Delivery Details")
        print("Customer:", self.customer_name)
        print("Address:", self.address)

# taking input from user
name = input("Customer Name: ")    
address = input("Address: ")       
# creating object 
d1 = Delivery(name, address)
# calling method to display delivery details
d1.print_details()
```
