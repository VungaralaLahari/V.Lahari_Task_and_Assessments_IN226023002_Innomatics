# Problem Statement 1: Smart Parking Lot Management System


```python
def smart_parking_system(parking_capacity, vehicle_logs):
    # Tracks current number of vehicles inside parking
    current_vehicles = 0
    # Tracks highest number of vehicles parked
    peak_vehicles = 0

    for log in vehicle_logs:
        # Convert input to uppercase to handle case-insensitivity
        log = log.upper()

        if log == 'IN':
            if current_vehicles >= parking_capacity:
                print("Alert: Parking Full! Cannot allow more vehicles.")
                continue
            current_vehicles += 1

        elif log == 'OUT':
            # Prevent negative vehicle count
            if current_vehicles == 0:
                print("Alert: No vehicles to exit.")
                continue
            current_vehicles -= 1

        # Update peak inside loop
        if current_vehicles > peak_vehicles:
            peak_vehicles = current_vehicles
    
    # Display 
    print("Currently Parked Vehicles:", current_vehicles)
    print("Peak Parked Vehicles:", peak_vehicles)

    # Display the Parking status 
    if current_vehicles >= parking_capacity:
        print("Parking Status: Unavailable!")
    else:
        print("Parking Status: Available")

# Take inputs from User
parking_capacity = int(input("Enter Parking Capacity: "))
vehicle_logs = input("Enter Vehicle Entry & Exit (IN or OUT): ").split()

# Function Call
smart_parking_system(parking_capacity, vehicle_logs)
```

# Problem Statement 2: Online Food Delivery Time Estimator


```python
def time_estimator(distance, traffic_level, weather):
    # Base time per km
    time_per_km = 5
    base_time = distance * time_per_km

    traffic_delay = 0
    weather_delay = 0

    # Traffic condition
    traffic_level = traffic_level.lower()
    if traffic_level == 'high':
        traffic_delay += 20
    elif traffic_level == 'medium':
        traffic_delay += 10
    elif traffic_level == 'low':
        traffic_delay += 0
    else:
        print("Invalid traffic input")

    # Weather condition
    weather = weather.lower()
    if weather == 'rainy':
        weather_delay += 15
    elif weather == 'storm':
        weather_delay += 30
    else:
        weather_delay += 0

    # Final ETA
    final_time = base_time + traffic_delay + weather_delay
    print(f"Estimated Delivery Time: {final_time} minutes")


# Inputs
distance = int(input("Enter Distance (in kms): "))
traffic_level = input("Enter Traffic Level (high/medium/low): ")
weather = input("Enter Weather Condition (rainy/storm): ")

time_estimator(distance, traffic_level, weather)
```

# Problem Statement 3: Movie Theatre Seat Occupancy 

```python
def seat_occupancy(total_seats, booked_seats):
    # Count total booked seats
    booked_count = len(booked_seats)
    # Validate logical condition
    if booked_count > total_seats:
        print("Error: Booked seats exceed total capacity.")
        return
    # Calculate occupancy percentage
    occupancy = (booked_count / total_seats) * 100

    print(f"Occupancy: {occupancy:.2f}%")

    # Determine show status
    if occupancy == 100:
        print("Show Status: HOUSEFULL!!!")
    elif occupancy >= 80:
        print("Show Status: Almost Full")
    elif occupancy >= 50:
        print("Show Status: Half Full")
    elif occupancy >= 20:
        print("Show Status: Almost Empty")
    else:
        print("Show Status: Empty")


total_seats = int(input("Enter Total Number of Seats: "))
booked_seats = list(map(int, input("Enter Bookings: ").split()))
seat_occupancy(total_seats, booked_seats)
```

# Problem Statement 4: Cloud Server Load Classification System


```python
def server_classifier(cpu_readings): 
    # Validate empty list
    if len(cpu_readings) == 0:
        print("No CPU readings provided.")
        return
    # Calculate average CPU usage
    avg_cpu = sum(cpu_readings) / len(cpu_readings)
    print(f"Average CPU Load: {avg_cpu:.2f}%")

    # Determine server status
    if avg_cpu < 50:
        print("Server Status: Normal")
    elif avg_cpu <= 80:
        print("Server Status: Warning")
    else:
        print("Server Status: Critical")


# Input from user
cpu_readings = list(map(int, input("Enter CPU Readings: ").split()))
# Function call
server_classifier(cpu_readings)
```

# Problem Statement 5: Smart Classroom Resource Usage Monitor


```python
def classroom_resource_monitor(resource_usage):
    
    # Create an empty list to store overused resources
    overused_resources = []
    
    # Loop through each resource and its usage hours
    for resource, hours in resource_usage.items():
        
        # Check if usage is more than 8 hours
        if hours > 8:
            overused_resources.append(resource)
    
    # Display overused resources
    if len(overused_resources) > 0:
        print("Overused Resources:", ", ".join(overused_resources))
        print("Energy Alert: Yes")
    else:
        print("Overused Resources: None")
        print("Energy Alert: No")


# Sample Input
resource_usage = {
    "Projector": 6,
    "AC": 9,
    "Lights": 4
}

# Function Call
classroom_resource_monitor(resource_usage)
```

#  Problem Statement 6: Online Event Registration Capacity Controller


```python
def manage_event_registration(event_capacity, total_registrations):
    
    # If registrations are less than or equal to capacity
    if total_registrations <= event_capacity:
        confirmed = total_registrations
        waitlisted = 0
        status = "Open"
    
    # If registrations exceed capacity
    else:
        confirmed = event_capacity
        waitlisted = total_registrations - event_capacity
        status = "Closed"
    
    # Display results
    print("Confirmed Registrations:", confirmed)
    print("Waitlisted Users:", waitlisted)
    print("Registration Status:", status)


# Take input from user
event_capacity = int(input("Enter Event Capacity: "))
total_registrations = int(input("Enter Total Registrations: "))

# Function call
manage_event_registration(event_capacity, total_registrations)
```
