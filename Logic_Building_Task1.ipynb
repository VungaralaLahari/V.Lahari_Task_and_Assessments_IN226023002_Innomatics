### Question 1
Verify the username and password and return the output accordingly

```python
# give username and password
username_given="admin"
password_given="1234"

# taking input from the user 
username=input("Enter your Username:")
password=input("Enter Password:")

# checking or verifying 
if username==username_given and password==password_given:
    print("Login Successful!")
else:
    print("Invalid Credentials")
    
    
```

### Question 2
Pass or Fail analyser 

```python
# lets take the data first 
marks=[45,78,90,33,60]
pass_count=0
fail_count=0
# running loop to access all elements of list
for mark in marks:
    # checking the condition
    if mark>=50:
        # increasing the count
        pass_count+=1
    else:
        fail_count+=1
# printing the output 
print("Total count of Passed Students: ",pass_count)
print("Total count of Failed Students: ",fail_count)
```

### Question 3
Data cleaner , data preprocessing before analysis 

```python
# data 
names=[" Alice ","bob"," CHARLIE "]

# empty list to add the cleaned data 
cleaned_names=[]

for name in names:
    # to remove space and convert to lowercase 
    name=name.strip().lower()
    # adding the elements to list 
    cleaned_names.append(name)

# displaying the data 
print("Cleaned Data: ",cleaned_names)
```

### Question 4
Message length Analyser 

```python
# data 
messages=["Hi","Welcome to the platform","OK"]
for message in messages:
    # checking the length
    if len(message)>10:
        # displaying the output and flagging issues 
        print(f"Length of '{message}' is {len(message)}-> its too long!")
    else:
        print(f"Length of '{message}' is {len(message)}-> its short")
```

### Question 5 
Error Message Detector 

```python
# data 
logs=['INFO','ERROR','WARNING','ERROR']
error_count=0
for i in logs :
    if i =='ERROR':
        error_count+=1
# displaying the count
print("Error Count: ",error_count)
```
